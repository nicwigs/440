```python
#########################
# DO NOT CHANGE!!!!!
##########################
# An instance of this class
# is what your code will
# be given.
#########################
class CountriesGraph:
    def __init__(self,graph_dict):
        # initialize from a dictionary
        self.__graph_dict = graph_dict
        
    def neighbors(self,country):
        # returns a list of neighbors
        return self.__graph_dict[country]
    
    def check_path(self,origin,destination,path):
        # origin and destination should be strings
        # path should be a list of strings
        if not path:
            # path can't be empty
            return False
        if path[0] == origin and path[-1] == destination:
            prev_node = path[0]
            # for every node in the path, check
            # that it is a neighbor of the previous
            # node
            for node in path[1:]:
                if node not in self.neighbors(prev_node):
                    return False
                prev_node = node
            return True
        return False


#######################################
# Change only code beneath this point!
########################################
'''
    Depth first search calling function
'''
def DFS(graph,origin,destination):
    path = []
    return rdfs(graph,origin,destination,path)
    
'''
    Recursive depth first search
'''
def rdfs(graph,current,destination,path):
    path.append(current)
    for country in graph.neighbors(current):
        # We are done- return. 
        # Places before following if, in case go back through destination, dont add again
        if destination in path:                     
            return path
        # Add destination if we are here
        if country == destination:
            path.append(country)
            return path
        # Path not found yet, keep expanding down
        if country not in path:
            path = rdfs(graph,country,destination,path)
    # ran out of nodes to explore, remove current, go back to parent node     
    if destination not in path:
        path.pop()
        
    return path
# finish this function!!
    
# Usage example:
# path =  DFS(euGraph,'Germany','Greece')
```
